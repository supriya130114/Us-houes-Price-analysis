```python
data['date'] = pd.to_datetime(data.date)
output['date'] = pd.to_datetime(output.date)

data['year'] = pd.DatetimeIndex(data.date).year
data['month'] = pd.DatetimeIndex(data.date).month
data['weekday'] = pd.DatetimeIndex(data.date).weekday
data['day'] = pd.DatetimeIndex(data.date).day

output['year'] = pd.DatetimeIndex(output.date).year
output['month'] = pd.DatetimeIndex(output.date).month
output['weekday'] = pd.DatetimeIndex(output.date).weekday
output['day'] = pd.DatetimeIndex(output.date).day
```

# **Visual Analysis**

```python
sns.pairplot(usa_homes);
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = usa_homes['Avg. Area Income'], y= usa_homes.Price)
plt.title('Relationship between Income and House Prices');
```

```python
with plt.xkcd():
  plt.figure(figsize=(12,6))
  sns.scatterplot(x = usa_homes['Area Population'], y= usa_homes.Price)
  plt.title('Relationship between Area Population and Home Prices ')
```

```python
with plt.xkcd():
  plt.figure(figsize=(12,6))
  sns.barplot(x = output.condition, y= output.price)
  plt.title('Relationship between House Condition and Prices');
```

```python
with plt.xkcd():

  fig, axes = plt.subplots(2, 2, figsize=(16, 8))
  
  axes[0,0].set_title('Number of Bedrooms Vs Prices')
  sns.barplot(x= output.bedrooms, y= output.price, data= output, ax= axes[0,0])

  axes[0,1].set_title('Number of Bathrooms Vs Prices')
  sns.barplot(x= output.bathrooms, y= output.price, data= output, ax= axes[0,1])

  axes[1,0].set_title('Size of Living Room Vs Prices')
  sns.scatterplot(x= output.sqft_living, y= output.price, data= output, ax= axes[1,0])

  axes[1,1].set_title('Number of Floors Vs Prices')
  sns.barplot(x= output.floors, y= output.price, data= output, ax= axes[1,1])

```

```python
with plt.xkcd():
  fig, axes = plt.subplots(2, 2, figsize=(16, 8))
  axes[0,0].set_title('Relationship between How old a house is and the price')
  sns.scatterplot(x= output.yr_built, y= output.price, data= output, ax= axes[0,0])

  axes[0,1].set_title('Relationship between House renovation and prices')
  sns.scatterplot(x= output.yr_renovated, y= output.price, data= output, ax= axes[0,1])

  axes[1,0].set_title('Does Number of Views affect House Prices?')
  sns.barplot(x= output.view, y= output.price, data= output, ax= axes[1,0])

  axes[1,1].set_title('Water Front vs Prices')
  sns.barplot(x= output.waterfront, y= output.price, data= output, ax= axes[1,1])

```
